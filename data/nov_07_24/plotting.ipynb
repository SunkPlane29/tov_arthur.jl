```python
# agora eu vou plotar as curvas massa raio 
plt.figure(figsize=(6,4))

mit_color= "#ef476f"

eta_13 = "#ffd166"
eta_14 = "#06d6a0"
eta_15 = "#118ab2"

# MIT BAG MODEL 
data1 = pd.read_csv("MR_diag_MIT.csv")
plt.plot(data1.M,data1.R,label="MIT $B^{1/4}=145$ MeV",color=mit_color)


# TRS
data1 = pd.read_csv("MR_diag_TRS_13.csv")
plt.plot(data1.M,data1.R,label="TRS $\eta=1.3$",ls='--',color=eta_13)

data1 = pd.read_csv("MR_diag_TRS_14.csv")
plt.plot(data1.M,data1.R,label="TRS $\eta=1.4$",ls='--',color=eta_14)

data1 = pd.read_csv("MR_diag_TRS_15.csv")
plt.plot(data1.M,data1.R,label="TRS $\eta=1.5$",ls='--',color=eta_15)


# MSS
data1 = pd.read_csv("MR_diag_MSS13.csv")
plt.plot(data1.M,data1.R,label="MSS $\eta=1.3$",ls='-',color=eta_13)

data1 = pd.read_csv("MR_diag_MSS14.csv")
plt.plot(data1.M,data1.R,label="MSS $\eta=1.4$",ls='-',color=eta_14)

data1 = pd.read_csv("MR_diag_MSS15.csv")
plt.plot(data1.M,data1.R,label="MSS $\eta=1.5$",ls='-',color=eta_15)

plt.xlabel("$R$[km]",fontsize=14)
plt.ylabel("$M$[$M_{\odot}$]",fontsize=14)

plt.ylim(0,5)
plt.xlim(5,30)

plt.grid(alpha=0.2)
plt.legend(fontsize=10,loc=2)
plt.savefig("mr_diagrams_07nov24.png",dpi=300)
plt.show()
```
